### Implementation of multiple variable regression.

```python
#define a function to get the no of input varible and no of target variable.

def n_x_y(X,Y):#should pass dataset of form m*n where m is no of instances n is no of vaiables
    n_x=X.shape[1]
    n_y=Y.shape[1]
    return n_x,n_y
```

```python
#initialize model parameters W and b

def init_params(n_x,n_y):
    W=np.random.randn(n_y, n_x)*0.01 #W shape ny*nx
    b=np.random.randn(n_y,1) #b shape n_y*1
    return W,b
```

```python
#forward propagation

def forward_prop(X,W,b):
    y_hat=np.matmul(W,X.T)+b #gives yhat in the shape ny*m
    return y_hat.T #m*ny
```

```python
#Computing cost function.
def cost(Y,Y_hat):
    return np.sum((Y_hat-Y)**2)/(2*(Y.shape[0]))
```

```python
#update parameters
def update_param(Y_hat,Y,X,W,b,learning_rate):
    slope_arr= (np.matmul((Y_hat-Y).T,X))/len(Y)
    W=W-learning_rate*slope_arr
    b = b - learning_rate * np.mean(Y_hat - Y, axis=0).reshape(-1, 1)
    return W,b
```

```python
#develop a perceptron

def perceptron(X,Y,learning_rate,iterations):
    n_x,n_y= n_x_y(X,Y) #Get no of x and y
    W,b=init_params(n_x,n_y)# Get W and b
    for i in range(iterations):
        Y_hat=forward_prop(X,W,b) #Get Y_hat m*n_y
        loss=cost(Y,Y_hat)
        if (i%100==0):
            print(f"loss at iteration {i} is {loss}")
        W,b=update_param(Y_hat,Y,X,W,b,learning_rate)

    return W,b
```

```python
def predict(X,Y,X_pred,learning_rate,iterations):
    #normalizing data
    x_mean=np.mean(X,axis=0)
    x_std=np.std(X,axis=0)
    X=(X-x_mean)/x_std

    y_mean=np.mean(Y,axis=0)
    y_std=np.std(Y,axis=0)
    Y=(Y-y_mean)/y_std

    X_pred=(X_pred-x_mean)/x_std

    #calling perceptron
    W,b=perceptron(X,Y,learning_rate,iterations)

    #predicting
    Y_pred=np.matmul(W,X_pred.T)+b
    Y_pred=Y_pred.T
    #denormalizing
    Y_pred=Y_pred*y_std+y_mean

    return Y_pred
```
